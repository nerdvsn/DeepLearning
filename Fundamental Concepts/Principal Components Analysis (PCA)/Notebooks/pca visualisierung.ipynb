```python
X = df.drop("subject", axis=1).drop("Activity", axis=1)
y = df["Activity"]
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s = StandardScaler()
X = s.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
```

```python
# Wir sagen, dass wir die 2 wichtigsten Teile der Daten wollen.
p = PCA(n_components=2)

# Wir schauen uns die Daten genau an und lernen daraus, um die wichtigsten Teile zu finden.
p.fit(X)

# Wir verÃ¤ndern die Daten, um sie einfacher zu machen, indem wir nur die wichtigsten Teile behalten.
X_transformed = p.transform(X)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
# X_transformed[:, 0]: Das ist die erste Hauptkomponente (oder Achse). Diese Achse reprÃ¤sentiert die wichtigste Richtung in unseren Daten nach der PCA-Transformation.
# X_transformed[:, 1]: Das ist die zweite Hauptkomponente (oder Achse). Diese Achse reprÃ¤sentiert die zweitwichtigste Richtung in unseren Daten nach der PCA-Transformation.

plt.scatter(X_transformed[:, 0], X_transformed[:, 1])
plt.show()
```

```python
X_transformed_filted = X_transformed[y == "STANDING", :]
plt.scatter(X_transformed_filted[:, 0], X_transformed_filted[:, 1])
plt.show()
```

```python
y.unique()
```

```python
plt.figure(figsize=(15,6))
for activity in y.unique():
    X_transformed_filtered = X_transformed[y == activity, :]
    plt.scatter(X_transformed_filtered[:, 0], X_transformed_filtered[:, 1], label=activity, s=6)
plt.legend()
plt.show()
```

```python
p = PCA(n_components=3)
p.fit(X)
X_transformed = p.transform(X)



%matplotlib notebook

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(figsize=(15, 6))
ax = fig.add_subplot(111, projection='3d')

for activity in y.unique():
    X_transformed_filtered = X_transformed[y == activity, :]
    ax.scatter(X_transformed_filtered[:, 0], X_transformed_filtered[:, 1], X_transformed_filtered[:, 2], label=activity, s=4)

plt.show()
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
plt.legend()
```
