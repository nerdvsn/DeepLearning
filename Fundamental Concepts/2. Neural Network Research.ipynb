# 1. Sigmoid-Funktion

```python
def sigmoid(x):
    """
    Sigmoid function
    """
    return 1/(1 + np.exp(-x))

def centerAxis(uses_negative=False):
    # http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot
    ax = plt.gca()
    ax.spines['left'].set_position('center')
    if uses_negative:
        ax.spines['bottom'].set_position('center')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
```

```python
x = np.arange(-10, 10, 0.01)

y = sigmoid(x)
```

```python
centerAxis()
plt.plot(x, y)
plt.show()
```

# 2. Loss

```python
def loss(y, y_pred):
    return np.mean(np.square(y - y_pred) / 2)
```

# 3. Forward-Pass

```python

# Parameter input und output
x = 1.0
y = 0


w = 2.0
b = -1.0
l_rate = 0.1
```

```python
# Berechne z (Linieare Kombination)
z = x * w + b
z
```

```python
# Aktivierung (Prediction y_pred)

y_pred = sigmoid(x)
y_pred
```

# 4. Loss berechnen

```python
loss = loss(y=y, y_pred=y_pred)
loss
```

# 5. Backward-Pass: Gradienten Berechnung

```python
# Schritt 1: dL/dy_pred = y_pred - y_true
dL_dy_pred = y_pred - y
print(f"dL/dy_pred = {dL_dy_pred:.3f}")

# Schritt 2: dy_pred/dz = y_pred * (1 - y_pred)
dy_pred_dz = y_pred * (1 - y_pred)
print(f"dy_pred/dz = {dy_pred_dz:.3f}")

# Schritt 3: dz/dw = x
dz_dw = x
print(f"dz/dw = {dz_dw:.3f}")

# Gesamter Gradient für w:
dL_dw = dL_dy_pred * dy_pred_dz * dz_dw
print(f"dL/dw = {dL_dw:.3f}")

# Gesamter Gradient für b (dz/db = 1)
dz_db = 1
dL_db = dL_dy_pred * dy_pred_dz * dz_db
print(f"dL/db = {dL_db:.3f}")
```

```python
# Gewichte aktualisieren
w = w - (l_rate * dL_dw)
b = b - l_rate * dL_db

print(f"This ist the new Bias: {b} and ths ist the new weight w: {w}")
```
