```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
X = data[:, 5:6]
y = target
```

```python
a = 15.; b = -70.
lx = np.arange(4,10)
lguess = a*lx + b

plt.plot(lx, lguess, c='red')
plt.scatter(X, y, marker='.', alpha=0.5, color='blue')
plt.xlim(0, 10) 
plt.ylim(0, 60)
plt.xlabel('Durchschnittl. Anzahl Räume')
plt.ylabel('Preis ($1000)')
```

```python
X.shape, y.shape
```

## Ordinary Least Square

```python
import sklearn.linear_model as lm
```

```python
lr = lm.LinearRegression()
```

```python
lr.fit(X,y)
```

```python
lr.score(X,y)
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(X)
```

```python
y_pred.shape
```

```python
plt.scatter(X, y, marker='.', alpha=0.5, color='blue')
plt.scatter(X, y_pred, marker='.', alpha=0.5, color='red')
plt.xlim(0, 10)
plt.ylim(0, 60)
plt.xlabel('Durchschnittl. Anzahl Räume')
plt.ylabel('Preis ($1000)')
```

## Stochastic Gradient Descent

```python
lr = lm.SGDRegressor()
lr.fit(X,y)
```

```python
lr.score(X,y)
```

```python
pred = lr.predict(X)
```

```python
plt.scatter(X, y, marker='.', alpha=0.5, color='blue')
plt.scatter(X, pred, marker='.', alpha=0.5, color='red')
plt.xlim(0, 10)
plt.ylim(0, 60)
plt.xlabel('Durchschnittl. Anzahl Räume')
plt.ylabel('Preis ($1000)')
```

```python
"""
max_iter: Die maximale Anzahl der Iterationen (Epochen), 
die der SGD-Algorithmus durchführt. Wenn die maximale Anzahl erreicht ist, 
wird der Algorithmus gestoppt, unabhängig davon, ob er konvergiert ist oder nicht.

"""
lr = lm.SGDRegressor(max_iter=1000)
```

```python
lr.fit(X,y)
```

```python
lr.score(X,y)
```

```python
pred = lr.predict(X)
plt.scatter(X, y, marker='.', alpha=0.5, color='blue')
plt.scatter(X, pred, marker='.', alpha=0.5, color='red')
plt.xlim(0, 10)
plt.ylim(0, 60)
plt.xlabel('Durchschnittl. Anzahl Räume')
plt.ylabel('Preis ($1000)')
```

```python
# n_iter wird nicht mehr unterstützt, dies ist das Äquivalent 

lr = lm.SGDRegressor(tol=None)

lr.fit(X, y)
```

```python
lr.n_iter_
```

```python
lr.score(X, y)
```

```python
pred = lr.predict(X)
plt.scatter(X, y, marker='.', alpha=0.5, color='blue')
plt.scatter(X, pred, marker='.', alpha=0.5, color='red')
plt.xlim(0, 10)
plt.ylim(0, 60)
plt.xlabel('Durchschnittl. Anzahl Räume')
plt.ylabel('Preis ($1000)')
```
