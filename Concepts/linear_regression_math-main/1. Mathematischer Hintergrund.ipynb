```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Ordinary Least Squares (OLS)


```python
# Meine Messdaten

x_coords = np.array([1,2,3,4,5,6])
y_coords = np.array([3,5,7,8,12,15])

X =  np.stack([np.ones_like(x_coords), x_coords],).T
y = y_coords

# Pseudoinverse definieren
Xinv = np.linalg.inv(X.T @ X) @ X.T

# Berechnung der Modelparaneter
beta = Xinv @ y

# Extrahieren der Intercept- und Slope-Parameter
intercept, slope = beta

print(f"Intercept: {intercept}")
print(f"Slobe: {slope}")

# Plotten 
plt.scatter(x_coords, y_coords, marker='.', alpha=0.5, color='blue', label="Datenpunkte")

# intercept + slope * x_coords => b + axi
plt.plot(x_coords, intercept + slope * x_coords, color='red', label="Lineare Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

## Stochastic Gradient Descent (SGD) 

```python
x_coords = np.array([1,2,3,4,5,6])
y_coords = np.array([3,5,7,8,12,15])

# Initialisieren der Modellparameter
intercept = 0.0
slope = 0.0

# Lernrate (Steuert die Schrittgröße)
learning_rate = 0.1

# Anzahl der Iterationen
num_iterations = 1000

# SGD-Optimierungsschleife
for iteration in range(num_iterations):
    # Berechnung der Vorhersage
    y_pred = intercept + slope * x_coords 
    
    # Berechnung des Fehlers
    error = y_coords - y_pred
    
    # Aktualisierung der Parameter
    intercept += learning_rate * np.mean(error)
    slope += learning_rate * np.mean(error * x_coords)

# Ausgabe der gefundenen Parameter
print("Intercept (b0):", intercept)
print("Slope (b1):", slope)

# Plot der Daten und der linearen Regression
plt.scatter(x_coords, y_coords, label="Datenpunkte")
plt.plot(x_coords, intercept + slope * x_coords , color='red', label="Lineare Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
